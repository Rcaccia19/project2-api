```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags

if "link" not in scripts.columns:
    raise ValueError("The CSV file must have a 'link' column.")

script_texts = []


for index, row in scripts.iterrows():
    url = row["link"]
    
    try:
        
        response = requests.get(url, timeout=10)
        response.raise_for_status()  
       
        soup = BeautifulSoup(response.text, "html.parser")
        
        text_content = soup.get_text()

        
        trimmed_text = text_content.strip()[:1000]
    
    except Exception as e:
        trimmed_text = f"Error fetching {url}: {e}"
    
   
    script_texts.append(trimmed_text)

scripts["script_text"] = script_texts


scripts.to_csv("scripts_output.csv", index=False, encoding="utf-8")

print("Download complete. Data saved to 'scripts_output.csv'.")
```

```python
#download the movie posters for 10 of these movies and save them to this repository
for movie in movies[:10]:
    poster_path = movie['poster_path']
    poster_url = f"https://image.tmdb.org/t/p/w500{poster_path}"
    poster_data = requests.get(poster_url).content
    with open(f"poster_{movie['id']}.jpg", 'wb') as handler:
        handler.write(poster_data)
    print(f"Poster for '{movie['title']}' downloaded and saved as 'poster_{movie['id']}.jpg'")
```
